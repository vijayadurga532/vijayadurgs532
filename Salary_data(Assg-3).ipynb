```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
salary_path="F:\Projects\Data\Salary_Data.csv"
salary_data=pd.read_csv(salary_path)
```

```python
salary_data.head()
```

# LINE CHART

```python
plt.figure(figsize=(16,6))
sns.lineplot(data=salary_data)
```

SCATTER PLOT

```python
sns.scatterplot(x=salary_data['YearsExperience'], y=salary_data['Salary'])
```

BAR GRAPH

```python
plt.figure(figsize=(10,6))
plt.title("SALARY DATA")
sns.barplot(x=salary_data['YearsExperience'], y=salary_data['Salary'])
```

HISTOGRAM

```python
sns.histplot(salary_data['YearsExperience'])
```

COUNT PLOT

```python
plt.figure(figsize=(14,7))
sns.countplot(x='YearsExperience',data=salary_data)
```

HEATMAP

```python
plt.figure(figsize=(14,7))
plt.title("SALARY DATA")
sns.heatmap(data=salary_data,annot=True)
```

BOX PLOT

```python
plt.figure(figsize=(14,7))
sns.boxplot(x='YearsExperience',y='Salary',data=salary_data)
```

PAIR PLOT

```python
sns.pairplot(salary_data)
```

JOINT PLOT

```python
sns.jointplot(x='YearsExperience',y='Salary',data=salary_data)
```

DENSITY PLOT

```python
sns.kdeplot(data=salary_data['Salary'], shade=True)
```

```python
sns.kdeplot(data=salary_data['YearsExperience'], shade=True)
```

```python
sns.jointplot(x=salary_data['Salary'],y=salary_data['YearsExperience'],kind='kde')
```
