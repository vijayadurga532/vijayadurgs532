```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
startups_path="F:\Projects\Data\Startups.csv"
```

```python
startups_data=pd.read_csv(startups_path)
```

```python
startups_data.head()
```

LINE CHART

```python
plt.figure(figsize=(16,6))
sns.lineplot(data=startups_data)
```

SCATTER PLOT

```python
plt.figure(figsize=(14,6))
sns.scatterplot(data=startups_data)
```

BAR GRAPH

```python
plt.figure(figsize=(10,6))
plt.title("STARTUPS DATA")
sns.barplot(data=startups_data)
```

HISTOGRAM

```python
sns.histplot(data=startups_data)
```

COUNT PLOT

```python
plt.figure(figsize=(14,7))
sns.countplot(data=startups_data)
```

BOX PLOT

```python
plt.figure(figsize=(14,7))
sns.boxplot(data=startups_data)
```

PAIR PLOT

```python
sns.pairplot(startups_data)
```

JOINT PLOT

```python
sns.jointplot(data=startups_data)
```

DENSITY PLOT

```python
sns.kdeplot(data=startups_data, shade=True)
```
